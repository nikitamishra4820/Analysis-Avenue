20. Classify Movies Based on Ratings [Excellent, Good, and Average]


```python
def rating_cat(rating):
    if rating>7.0:
        return "Excellent"
    elif rating>=6.0:
        return "Good"
    else:
        return "Average"
```

```python
data['Category']=data['Rating'].apply(rating_cat)
```

22. Find Unique Values From Genre 


```python
data['Genre']
```

```python
list1 = []
for value in data['Genre']:
    list1.append(value.split(','))
```

```python
list1
```

```python
one_dlist=[]
for item in list1:
    for item1 in item:
        one_dlist.append(item1)
```

```python
one_dlist
```

```python
unique_list=[]
for item in one_dlist:
    if item not in unique_list:
        unique_list.append(item)
```

```python
unique_list
```
